# Attractor recovery score plot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../output/attractor_recovery.tsv', sep = '\t')

fraction_cols = ['att_recovery_score_IMR', 'att_recovery_score_BMR']
df_renamed = df.rename(columns={'att_recovery_score_IMR': 'IMR', 'att_recovery_score_BMR': 'BMR'})
df_renamed.set_index('model_no', inplace=True)
df_renamed = df_renamed.sort_values(by=['tot_nodes', 'max_indeg', 'mean_indeg'])

ytick_labels = [f"{idx}  | μ={round(mean, 2)}" for idx, mean in zip(df_renamed.index, df_renamed['mean_indeg'])]
ytick_labels = [
    f"{idx}  | $N={N}$, $k_{{\mathrm{{max}}}}={k_max}$, <k>={round(mean, 2)}"
    for idx, N, k_max, mean in zip(df_renamed.index, df_renamed['tot_nodes'], df_renamed['max_indeg'], df_renamed['mean_indeg'])
]

plt.figure(figsize=(8, 10))

ax = sns.heatmap(df_renamed[['IMR', 'BMR']], annot=True, cmap="Blues", vmin=0, vmax=1,
                 cbar_kws={"label": "Attractor recovery score"})

ax.set_yticklabels(ytick_labels, rotation=0, fontsize=8)
plt.ylabel("Topology of the networks")
plt.tight_layout()
#plt.savefig('../output/att_recovery_score.pdf', dpi = 600)
plt.show()
```

# Basin Recovery score plot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../output/basin_recovery.tsv', sep = '\t')

fraction_cols = ['ising_maj_basin_reco_score', 'null_maj_basin_reco_score']
df_renamed = df.rename(columns={'ising_maj_basin_reco_score': 'IMR', 'null_maj_basin_reco_score': 'BMR'})
df_renamed.set_index('model_no', inplace=True)
df_renamed = df_renamed.sort_values(by=['tot_nodes', 'max_indeg', 'mean_indeg'])

ytick_labels = [f"{idx}  | μ={round(mean, 2)}" for idx, mean in zip(df_renamed.index, df_renamed['mean_indeg'])]
ytick_labels = [
    f"{idx}  | $N={N}$, $k_{{\mathrm{{max}}}}={k_max}$, <k>={round(mean, 2)}"
    for idx, N, k_max, mean in zip(df_renamed.index, df_renamed['tot_nodes'], df_renamed['max_indeg'], df_renamed['mean_indeg'])
]

plt.figure(figsize=(8, 10))

ax = sns.heatmap(df_renamed[['IMR', 'BMR']], annot=True, cmap="Blues", vmin=0, vmax=1,
                 cbar_kws={"label": "Basin recovery score"})

ax.set_yticklabels(ytick_labels, rotation=0, fontsize=8)
plt.ylabel("Topology of the networks")
plt.tight_layout()
plt.savefig('../output/basin_recovery_score.png', dpi = 600)
plt.show()
```

# Plotting Basin recovery score as JS distance

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('../output/basin_recovery_JS_distance.tsv', sep = '\t')

# exclude_models = [61, 69, 95, 212]
# df = df[~df['model_number'].isin(exclude_models)]

fraction_cols = ['bio_gold_std_JS_dist', 'ising_gold_std_JS_dist', 'bit_gold_std_JS_dist']
df_renamed = df.rename(columns={'bio_gold_std_JS_dist': 'Published model' ,'ising_gold_std_JS_dist': 'IMR', 'bit_gold_std_JS_dist': 'BMR'})
df_renamed.set_index('model_no', inplace=True)
df_renamed = df_renamed.sort_values(by=['tot_nodes', 'max_indeg', 'mean_indeg'])

ytick_labels = [f"{idx}  | μ={round(mean, 2)}" for idx, mean in zip(df_renamed.index, df_renamed['mean_indeg'])]
ytick_labels = [
    f"{idx}  | $N={N}$, $k_{{\mathrm{{max}}}}={k_max}$, <k>={round(mean, 2)}"
    for idx, N, k_max, mean in zip(df_renamed.index, df_renamed['tot_nodes'], df_renamed['max_indeg'], df_renamed['mean_indeg'])
]

plt.figure(figsize=(8, 10))
#sns.heatmap(df_renamed[['Ising majority rule', 'Null majority rule']], annot=True, cmap="Blues", vmin=0, vmax=1, cbar_kws={"label": "Jaccard Similarity"})

ax = sns.heatmap(df_renamed[['Published model', 'IMR', 'BMR']], annot=True, cmap="Blues_r", vmin=0, vmax=1,
                 cbar_kws={"label": "JS distance from the gold standard"})

ax.set_yticklabels(ytick_labels, rotation=0, fontsize=8)
#ax.set_ylabel("Model Number  |  Mean Inputs")
#plt.title("Recovery of original fixed points by majority threshold rules")
plt.ylabel("Topology of the networks")
plt.tight_layout()
plt.savefig('../output/basin_recovery_score_JS.pdf', dpi = 600)
#plt.show()
```
